```python
search_list = ["devops engineer",
               "cloud engineer", 
               "devops consultant",
               "Site Reliability Engineer"]
url = "https://de.indeed.com/"
```

```python
def create_browser():
    options = Options()
    driver = webdriver.Chrome(options=options)
    return driver
driver=create_browser()    
```

```python
def goto_main_main_page(driver=None,url=None):
    try: 
        driver.get(url)
    except Exception as e:
        print("error in goto_main_main_page ",e)
goto_main_main_page(url=url,driver=driver)
```

```python
def search_keyword(keyword=None):
    #driver.find_element(By.ID,"text-input-what").clear()
    time.sleep(5)
    driver.find_element(By.ID,"text-input-what").send_keys(keyword)
    driver.find_element(By.ID,"text-input-what").send_keys(Keys.RETURN)
search_keyword(keyword="devops engineer")    
```

```python
#job_info = driver.find_element(By.CLASS_NAME,"jobsearch-InfoHeaderContainer.css-1toufe4.eu4oa1w0")
def find_job_title(div=None):
    try:
        job_title = div.find_element(By.CLASS_NAME,"jobsearch-JobInfoHeader-title.css-161nklr.e1tiznh50").text.replace("- job post","")
        return job_title
    except Exception as e:
        print("error in find_job_title >>",e)



def find_company_indeed_url(div=None):
    elements = ["css-1saizt3.e1wnkr790","css-1cxc9zk.e1wnkr790"]
    for el in elements:
        try:
            company_url_a_tag = div.find_element(By.CLASS_NAME,el)
            company_url = company_url_a_tag.find_element(By.TAG_NAME,'a').get_attribute('href')
            return company_url
        except Exception as e:
            pass#print("error[!] in find_company_url >> ",e)


def fetch_company_url(company_indeed_url,driver):
    try:
        website_driver.get(company_indeed_url)
        wait_the_company_link(driver=driver)
        company_url = driver.find_element(By.CSS_SELECTOR,'[class="css-1rezcpd e19afand0"]').get_attribute('href')
        return company_url
    except Exception as e:
        print("error in fetch_company_url")
    
    


    
def find_company_name(div=None):
    elements = ["css-1saizt3.e1wnkr790","css-1cxc9zk.e1wnkr790"]
    for el in elements:
        try:
            company_name = div.find_element(By.CLASS_NAME,el).text
            return company_name
        except Exception as e:
            pass#print("error[!] in find_company_name >> ",e)
def find_job_country(div=None):
    try:
        job_country = div.find_element(By.CSS_SELECTOR,'[data-testid="inlineHeader-companyLocation"]').text
        return job_country
    except Exception as e:
        print("error in find_job_country >>",e)

def get_job_posting_details(div=None):
    try:
        job_posting_details  = div.find_element(By.CSS_SELECTOR,'[class="date"]').text.replace("Posted","").strip()
        return job_posting_details
    except Exception as e:
        print("error in get_job_posting_detail ",e)
        
def random_sleep(a,b):
    wait_time = random.uniform(a, b)
    time.sleep(wait_time)

def write_data_frame(job_title,company_url,company_name,job_country,job_posting_date,source):
    data = {'job_title': [job_title],
            'company_url': [company_url],
             
            'company_name': [company_name],
            'job_country':[job_country],
            
            'job_posting_date':job_posting_date,
           'source': [source]}
    
    df = pd.DataFrame(data)
    csv_file_path = 'scraped_data/sample_output.csv'
    df.to_csv(csv_file_path, index=False,mode='a',header=False)

def wait_element_load():
    max_wait_time = 200
    
    try:
        element = WebDriverWait(driver, max_wait_time).until(
            EC.presence_of_element_located((By.ID,"jobsearch-ViewjobPaneWrapper"))
        )
    
    
    except:
        pass
def check_next_page():
    flag = False
    try:
        driver.find_element(By.CSS_SELECTOR,'[data-testid="pagination-page-next"]')
        flag = True 
    except Exception as e:
        print("error[!] in check_next_page")
    finally:
        return flag
def go_to_next_page(jobs_on_page):
    try:
        driver.find_element(By.CSS_SELECTOR,'[data-testid="pagination-page-next"]').click()

    except Exception as e:
        print("error[!] in go_to_next_page",e)

def close_pop_up():
    try:
        driver.find_element(By.CSS_SELECTOR,'[class="css-yi9ndv e8ju0x51"]').click()
    except Exception as e:
        pass


def wait_the_company_link(driver):
    max_wait_time = 20
    
    try:
        element = WebDriverWait(driver, max_wait_time).until(
            EC.presence_of_element_located((By.CSS_SELECTOR,'[class="css-1rezcpd e19afand0"]'))
        )
    
    except:
        pass

def get_the_number(input_string):
    numbers = re.findall(r'\d+', input_string)
    numbers = list(map(int, numbers))
    if numbers:
        return numbers[0]
    else:
        return 0


def format_the_job_publish_details(job_posting_details):
    print("job_posting_details >> ",job_posting_details)
    flag= False
    days =  get_the_number(input_string=job_posting_details)
    tz = pytz.timezone('Europe/Berlin')
    berlin_now = datetime.now(tz)

    if "Heute" in job_posting_details:
        job_post_date = berlin_now
    elif days < 29:
        job_post_date = berlin_now - timedelta(days=days)
    elif days > 29 and "Vor > 30 Tagen geschaltet" in job_posting_details:
         job_post_date =berlin_now-timedelta(days=30) 
         flag = True
    
    if flag:
        job_pusblish_date  = "before " + str(job_post_date.strftime('%m-%d-%Y')) 
    else:
        job_pusblish_date = job_post_date.strftime('%m-%d-%Y')
    return job_pusblish_date
```
