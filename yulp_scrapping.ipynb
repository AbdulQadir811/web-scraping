```python
from bs4 import BeautifulSoup
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
import re
import requests


url="https://www.yelp.com/search?find_desc=Restaurants&find_loc=San+Francisco%2C+CA"
#driver = webdriver.Chrome(ChromeDriverManager().install())

resturants_links = []
flag = True
while (flag ==True):
    print("url >>", url)
    driver.get(url)
    #soup=BeautifulSoup(driver.page_source)
    spans=driver.find_elements_by_class_name("css-uvzfg9")
    for span in spans:
        check = re.search("\A[0-9]",span.text)
        if check:
          
          print(span.text)
          link = span.find_element_by_tag_name('a')
          href_link = link.get_attribute('href')
          print(href_link)
          resturants_links.append(href_link)
    #break
    try:
        url = (driver.find_element_by_xpath("//*[@id='main-content']/div/ul/li[22]/div/div[1]/div/div[11]/span/a").get_attribute('href'))
    except Exception as e:
        print("end of the resturants !!!")
        flag = False
        
        

    
```

```python
import time 
import pandas as pd

for i,link in enumerate(resturants_links):
    print(">>>> ",i+1," ",link)
    my_dict = {}
    driver.get(link) 
    location_list = []
    phone_number_list = []
    resturant_website_list = []
    
    
    
    try:
        phone_number = driver.find_element_by_xpath("/html/body/yelp-react-root/div[1]/div[4]/div/div/div[2]/div/div[2]/div/div/section[1]/div/div[2]/div/div[1]/p[2]")
        phone_number = phone_number.text
        phone_number_list.append(phone_number) 
        my_dict["phone_number"] = phone_number_list
        
        
    except Exception as e:
        my_dict["phone_number"] = [""]
        print("[!] phone number not found")
        
        
    try:
        website = driver.find_element_by_xpath("/html/body/yelp-react-root/div[1]/div[4]/div/div/div[2]/div/div[2]/div/div/section[1]/div/div[1]/div/div[1]/p[2]/a")
        #print("link ",i,">>",website.text,"\n <<<<<<<<>>>>>>>>")
        resturant_website = website.text
        resturant_website_list.append(resturant_website)
        
        my_dict["website"] = resturant_website_list
        
    except Exception as e:
        my_dict["website"] = [""]
        print("[!] website not found")
        pass
    
    
    try:
        x = driver.find_element_by_css_selector(".container__09f24__fZQnf.border-color--default__09f24__NPAKY.overflow--hidden__09f24___ayzG")
        y = x.find_element_by_tag_name("img")
        z = y.get_attribute("src")
        z = z.split("&center=")[1]
        z = z.split("&markers=")[0]
        z = z.split("%2C")
        z = [float(i) for i in z]
        #print("location",z)
        location  = str(z[0])+" "+str(z[1])
        location_list.append(location)
        
        my_dict["location"]  = location_list
        #my_dict["location"]  = (location)
        #location_list.append(location)
        #print(my_dict["location"])
    except Exception as e:
        my_dict["location"] = [""]
        #location_list.append("")
        print(e)
        print("[!] Location Not Found")
        
    try:
        print(my_dict)
        df = pd.DataFrame(my_dict) 
        df.to_csv ('test18.csv',mode='a',index=False, header=False)
        print("<<<<<<<<record edit>>>>>")
    except Exception as e:
        print(e)
    
```

```python
# Python program to scrape table from website

# import libraries selenium and time
from selenium import webdriver
from time import sleep

# Create webdriver object

# Get the website
driver.get(
	"https://www.geeksforgeeks.org/find_element_by_link_text-driver-method-selenium-python/")

# Make Python sleep for some time
sleep(2)

# Obtain the number of rows in body
rows = 1+len(driver.find_elements_by_xpath(
	"/html/body/div[3]/div[2]/div/div[1]/div/div/div/article/div[3]/div/table/tbody/tr"))

# Obtain the number of columns in table
cols = len(driver.find_elements_by_xpath(
	"/html/body/div[3]/div[2]/div/div[1]/div/div/div/article/div[3]/div/table/tbody/tr[1]/td"))

# Print rows and columns
print(rows)
print(cols)

# Printing the table headers
print("Locators		 "+"			 Description")

# Printing the data of the table
for r in range(2, rows+1):
	for p in range(1, cols+1):
		
		# obtaining the text from each column of the table
		value = driver.find_element_by_xpath(
			"/html/body/div[3]/div[2]/div/div[1]/div/div/div/article/div[3]/div/table/tbody/tr["+str(r)+"]/td["+str(p)+"]").text
		print(value, end='	 ')
	print()
```
